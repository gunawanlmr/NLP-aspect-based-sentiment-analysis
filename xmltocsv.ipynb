```python
import pandas as pd
```

```python
import csv
import xml.etree.ElementTree as et
from collections import OrderedDict

doc = et.parse('dataset/data/training_set.xml')
field = ['rid', 'text', 'aspect1_food', 'aspect1_ambiance', 'aspect1_price', 'aspect1_service', 'aspect2_food', 'aspect2_ambiance', 'aspect2_price', 'aspect2_service']
data = pd.DataFrame(columns=field)
row_list = []

def update_aspect(aspects):
    aspect = aspects.findall('aspect')
    aspect_food = None
    aspect_ambiance = None
    aspect_price = None
    aspect_service = None
    for item in aspect:
        cat = item.attrib.get('category')
        if cat == 'FOOD':
            aspect_food = item.attrib.get('polarity')
        if cat == 'AMBIENCE':
            aspect_ambiance = item.attrib.get('polarity')
        if cat == 'PRICE':
            aspect_price = item.attrib.get('polarity')
        if cat == 'SERVICE':
            aspect_service = item.attrib.get('polarity')
            
    return aspect_food, aspect_ambiance, aspect_price, aspect_service

for review in doc.findall('review'):
    review_dict = {}
    rid = review.attrib.get('rid')
    text = review.find('text').text
    aspects = review.findall('aspects')
    aspect2_food = None
    aspect2_ambiance = None
    aspect2_price = None
    aspect2_service = None
    
    aspect1_food, aspect1_ambiance, aspect1_price, aspect1_service = update_aspect(aspects[0])
    if len(aspects) == 2:
        aspect2_food, aspect2_ambiance, aspect2_price, aspect2_service = update_aspect(aspects[1])
    
    review_dict['rid'] = rid
    review_dict['text'] = text
    review_dict['aspect1_food'] = aspect1_food
    review_dict['aspect1_ambiance'] = aspect1_ambiance
    review_dict['aspect1_price'] = aspect1_price
    review_dict['aspect1_service'] = aspect1_service
    review_dict['aspect2_food'] = aspect2_food
    review_dict['aspect2_ambiance'] = aspect2_ambiance
    review_dict['aspect2_price'] = aspect2_price
    review_dict['aspect2_service'] = aspect2_service
    
    row_list.append(review_dict)

data = df = pd.DataFrame(row_list)  
    
```
